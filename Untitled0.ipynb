```python
import pandas as pd;
```

```python
data=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")

def clean(data, is_test=False):
  if is_test:
    data = data.drop(["Ticket","Cabin","Name"], axis=1)
  else:
    data = data.drop(["Ticket","Cabin","Name","PassengerId"], axis=1)

  cols=["SibSp","Parch","Fare","Age"]
  for col in cols:
    data[col] = data[col].fillna(data[col].median()) # Corrected fillna

  data["Embarked"] = data["Embarked"].fillna(data["Embarked"].mode()[0])
  return data

data = clean(data) # Call the clean function on the training data
test = clean(test, is_test=True) # Call the clean function on the test data and specify it's the test set
```

```python
data.head(5)
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
cols=["Sex","Embarked"]
for col in cols:
  data[col]=le.fit_transform(data[col])
  test[col]=le.transform(test[col])
  print(le.classes_)



data.head(5)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

y=data["Survived"]
x=data.drop("Survived",axis=1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LogisticRegression(max_iter=200)
model.fit(x_train, y_train)

pred = model.predict(x_test)
```

```python
predict=model.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predict))
```

```python
# Make predictions on the actual test data
test_predictions = model.predict(test.drop("PassengerId", axis=1))

# Create the submission DataFrame
df = pd.DataFrame({"PassengerId": test["PassengerId"], "Survived": test_predictions})

# Save the submission file
df.to_csv("submission.csv", index=False)

print("Submission file 'submission.csv' created successfully!")
```
